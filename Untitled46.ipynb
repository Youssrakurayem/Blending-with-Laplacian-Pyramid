```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
#generating laplacian pyramid by taking two inputs the image itself and the number of the levels needed
def generate_laplacian_pyramid(image, levels):
  #list that will hold the gaussian pyramid
    gaussian_pyramid = [image]
    for _ in range(levels - 1):
      #the pyrDown method is used to downsample the image and smooth it and then add it to the list
        image = cv2.pyrDown(image)
        gaussian_pyramid.append(image)
    #here goes the creation of the laplacian pyramid that takes the last level in the gaussain pyramid
    laplacian_pyramid = [gaussian_pyramid[levels - 1]]
    #a loop in a reverse order and the pyrUp method is used to upsample the image and then applies Gaussian smoothing filter
    for i in range(levels - 1, 0, -1):
        expanded = cv2.pyrUp(gaussian_pyramid[i])
     #if the size of the upsampled image "expanded" doesnt match the size of the next level in the gaussian then its resized to match
        if expanded.shape[:2] != gaussian_pyramid[i - 1].shape[:2]:
            expanded = cv2.resize(expanded, gaussian_pyramid[i - 1].shape[:2][::-1])
        #subtracting the expanded from the next level in the gaussian pyramid
        laplacian = cv2.subtract(gaussian_pyramid[i - 1], expanded)
        laplacian_pyramid.append(laplacian)

    return laplacian_pyramid
```

```python
#reconstruncting the image from laplacian pyramid list
def reconstruct_image_from_laplacian_pyramid(laplacian_pyramid):
  #starting by the first level in the pyramid
    reconstructed_image = laplacian_pyramid[0]
    #loop over the list
    for i in range(1, len(laplacian_pyramid)):
      #upsampling the image and applying gaussian filter
        expanded = cv2.pyrUp(reconstructed_image)
        #making sure that the sizes match
        if expanded.shape[:2] != laplacian_pyramid[i].shape[:2]:
            expanded = cv2.resize(expanded, laplacian_pyramid[i].shape[:2][::-1])
         #The upsampled image "expanded" and (laplacian_pyramid[i]) are added together using the cv2.add() function
        reconstructed_image = cv2.add(expanded, laplacian_pyramid[i])

    return reconstructed_image
```

```python
#taking image 1 and image 2 and the number of levels will be equal to six
def blend_images(image1, image2, levels=6):
  #calculating the width and the height of the image
    height, width = image1.shape[:2]
    #dividing the width by 2 to be able to extract half of the image
    half_width = width // 2

    image1_left_half = image1[:, :half_width]
    image2_right_half = image2[:, half_width:]

    laplacian_pyramid1_left = generate_laplacian_pyramid(image1_left_half, levels)
    laplacian_pyramid2_right = generate_laplacian_pyramid(image2_right_half, levels)

   #a list to store the blended levels of laplacian pyramids
    blended_pyramid = []
    #iterating over each level of the Laplacian pyramids of both images simultaneously using zip
    for (laplacian1_left, laplacian2_right) in zip(laplacian_pyramid1_left, laplacian_pyramid2_right):
      #taking the minimum number of the rows and columns to make sure the sizes match togheter
        min_rows = min(laplacian1_left.shape[0], laplacian2_right.shape[0])
        min_cols = min(laplacian1_left.shape[1], laplacian2_right.shape[1])
        laplacian1_left = laplacian1_left[:min_rows, :min_cols]
        laplacian2_right = laplacian2_right[:min_rows, :min_cols]

        blended_level = np.hstack((laplacian1_left, laplacian2_right))
        blended_pyramid.append(blended_level)

    blended_image = reconstruct_image_from_laplacian_pyramid(blended_pyramid)
    return blended_image
```

```python
def test_blend_images():
    image1 = cv2.imread('/content/WC202209_LAST_006_430968_reduced.jpg')
    image2 = cv2.imread('/content/334289821-Baltimore_Oriole-Matthew_Plante.jpg')

  #trying different levels
    for levels in [3, 5, 7]:
        blended_image = blend_images(image1, image2, levels)
        cv2_imshow(blended_image)

test_blend_images()
```
